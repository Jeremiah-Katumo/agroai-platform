```python
from fastai.vision.all import *
```

```python
from pathlib import Path
path = Path('/content/new_plant_diseases_data')

fns = get_image_files(path)
fns[:5]
```

```python
failed = verify_image(fns)
failed
```

```python
leaf = DataBlock(
    blocks=(ImageBlock, CategoryBlock),             # We're working with images and categories (labels)
    get_items=get_image_files,                      # Get all image files from the path
    splitter=RandomSplitter(valid_pct=0.2, seed=42),# Randomly split 20% for validation
    get_y=parent_label,                             # Get label from the folder name (e.g., 'Tomato___Late_blight')
    item_tfms=Resize(224)                           # Resize all images to 224x224
)
leaf
```

```python
dls = leaf.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
learner = vision_learner(dls, resnet18, metrics=error_rate)
learner.fine_tune(3)
```

```python
lr_min, lr_steep = learner.lr_find(suggest_funcs=(minimum, steep))
```

```python
lr_steep
```

```python
learner.fine_tune(18,base_lr=lr_steep,cbs=[ShowGraphCallback()])
```

```python
# Confusion matrix
interp = ClassificationInterpretation.from_learner(learner)
interp.plot_confusion_matrix(
    figsize=(20, 20),     # Increase size to spread out the labels
    dpi=100,              # Sharper resolution
    cmap='Blues',         # Brighter color map
    normalize=False,      # Optional: Normalize values
)
```

Visualize the top 5 losses to understand where the model is making the biggest mistakes to help in identifying noisy labels or images that are difficult to classify.

```python
interp.plot_top_losses(5, nrows=1, figsize=(15, 5))
```

```python
# Save Model
learner.export()
path = Path()
path.ls(file_exts='.pkl')
```

```python
# Load Model
learn_inf = load_learner(path/'export.pkl')
```

```python
learn_inf.dls.vocab
```

```python
test_path = Path('/content/new_plant_diseases_data/test/test')
test_files = get_image_files(test_path)
test_dl = dls.test_dl(test_files)
preds, _ = learn_inf.get_preds(dl=test_dl)
pred_classes = preds.argmax(dim=1)
labels = dls.vocab

for file, pred in zip(test_files[:10], pred_classes[:10]):
    print(f"{file.name}  {labels[pred]}")
```
